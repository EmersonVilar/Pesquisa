```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def step_signal (low, high, s_low=0.2, s_high=0.2, larg_high_i=20, larg_high_f=100, larg_low_i=20, larg_low_f=100, prob_positivo=0.5, n=100):
       
    distrib = np.random.binomial(1,prob_positivo,n)
    signal = np.concatenate([np.random.uniform(high-s_high,high+s_high,np.random.randint(larg_high_i, larg_high_f)) if x == 1 else np.random.uniform(low-s_low,low+s_low,np.random.randint(larg_low_i, larg_low_f)) for x in distrib])
    return signal
```

```python
def add(log_x,n,N):
    vetor_media = np.array(vect_random(log_x,n))
    media = []
    for i in np.arange(0,N):
        media.append(vetor_media.mean())
    
    media_td = np.array(media)
    mTd = media_td.mean()
    sTd = media_td.std()
    return mTd,sTd    
```

```python
def vect_random(signal,n):
    amostra = []
    for i in np.arange(0,n):
        amostra.append(signal[np.random.randint(0,len(signal))])
    return amostra
```

```python
signal = step_signal(3, 9, 2, 1)
plt.figure(figsize=(16,4))
plt.plot(signal)
plt.show()
```
