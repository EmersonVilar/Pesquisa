```python
import numpy as np
import matplotlib.pyplot as plt

def step_signal(low, s_low, high, s_high):
    
    distrib = np.random.binomial(1,0.5,100)
    signal = np.concatenate([np.random.uniform(high-s_high,high+s_high,np.random.randint(20,160)) if x == 1 else np.random.uniform(low-s_low,low+s_low,np.random.randint(20,160)) for x in distrib])
    return signal
    
# plt.figure(figsize=(16,4))
# plt.plot(signal)
# plt.show()
```

```python
signal = step_signal(3, 0.5, 9, 0.2)
```

```python
plt.figure(figsize=(16,4))
plt.plot(signal)
plt.show()
```
